```python
import pickle
import pandas as pd
from datetime import datetime
from pprint import pprint
import fa_mods as famods
import fa_utils

from collections import defaultdict

with open('/home/craigc/.keys/financialmodelingprep.txt', 'r') as file:
    api_key = file.read()
```

```python
symbol = 'MSFT'
data = fa_utils.get_ticker_data(symbol, api_key)
```

```python
# Print all of the names of all metrics gathered
metric_cnt = defaultdict(int)

for key in data.keys():
    val = data[key]
    if isinstance(val, pd.DataFrame):
        print(f'\n## {key}')

        if key == 'dividends' or key == 'scores':
            val = val.T

        for metric in val.index.values.tolist():
            metric_cnt[metric] += 1
            print(f'- {metric}')
```

```python
metrics_sorted_by_freq = sorted(metric_cnt.items(), key=lambda x: x[1], reverse=True)
pprint(metrics_sorted_by_freq)
```
